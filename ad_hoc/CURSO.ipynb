```python
from pyspark.sql import functions as f
```

```python
df_curso = (
    spark
    .read
    .options(header=True, sep='|')
    .csv('/home/rafael/Downloads/microdados_educacao_superior_2019/Microdados_Educaç╞o_Superior_2019/dados/SUP_CURSO_2019.CSV')
)
```

```python
df_curso.printSchema()
```

```python
df_ies = (
    spark
    .read
    .options(header=True, sep='|')
    .csv('/home/rafael/Downloads/microdados_educacao_superior_2019/Microdados_Educaç╞o_Superior_2019/dados/SUP_IES_2019.CSV')
)
```

```python
df_ies.printSchema()
```

```python
df = df_curso.join(df_ies, 'CO_IES', 'LEFT')
```

```python
df.createOrReplaceTempView('curso')
```

```python
spark.sql(
'''
    select count(distinct CO_CURSO), CO_REGIAO
    from curso
    where upper(NO_CURSO) = 'DIREITO'
    group by CO_REGIAO
'''
).toPandas().style
```

```python
spark.sql(
'''
    select sum(QT_INSCRITO_TOTAL) qtd, no_curso
    from curso
    group by no_curso 
    order by qtd desc
'''
).toPandas().style
```
